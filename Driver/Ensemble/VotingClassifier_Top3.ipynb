```python
import numpy as np
import matplotlib
import pandas as pd
from sklearn.model_selection import train_test_split


dataset = pd.read_csv('/Users/wzhang/Documents/injury data cleaned.csv')

X = pd.DataFrame(dataset.iloc[:,1:19].values)
Y = pd.DataFrame(dataset.iloc[:,19:].values)
print(X)
print(Y)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

print(X_train.shape)
print(Y_train.shape)
```

```python
from keras.models import Sequential
import tensorflow as tf
from tensorflow.keras.layers import LSTM, Dropout, Dense, BatchNormalization
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.neural_network import MLPClassifier
def create_mlp():
    model = MLPClassifier(max_iter=1000, tol=1e-4)
    return model
```

```python
from sklearn.linear_model import SGDClassifier
def create_SGD():
    model = SGDClassifier(max_iter=1000, tol=1e-4)
    return model
```

```python
from sklearn.ensemble import RandomForestClassifier
def create_rf():
    model = RandomForestClassifier(n_estimators=100)
    return model
```

```python
clf1 = create_mlp()
clf2 = create_SGD()
clf3 = create_rf()
```

```python
from sklearn.ensemble import VotingClassifier
model = VotingClassifier(estimators=[('MLP', clf1), ('SGD', clf2), ('RF', clf3)], n_jobs=10)
model.fit(X_train,Y_train)
```

```python
res = model.predict(X_test)
print(res)
print(Y_test)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(Y_test,res)
print(accuracy)
from sklearn.metrics import classification_report
print(classification_report(Y_test,res,digits=6))
```
